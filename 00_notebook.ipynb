```python
import pandas as pd
import os
from easyreg import register_images_mni_152, warp_image, img_to_2mm
inputs = pd.read_csv("input/inputs.csv")

display(inputs)

for i, row in inputs.iterrows():
    print("Working on ...", row["subject"])
    input_image = row["native_image"]
    input_roi = row["native_roi"]
    flo_seg = row["segmented_image"]
    flo_reg = row["registered_image"]
    fwd_field = row["fwd_field"]
    roi_reg = row["registered_roi"]
    roi_2mm = row["roi_2mm"]

    if not os.path.exists(roi_reg):
        register_images_mni_152(flo = input_image, 
                                flo_seg = flo_seg, 
                                flo_reg = flo_reg, 
                                fwd_field = fwd_field, 
                                threads=4)
        warp_image(input_roi, roi_reg, fwd_field)
    if not os.path.exists(roi_2mm):
        img_to_2mm(roi_reg, roi_2mm)
```
